```python
import xarray as xr
import matplotlib.pyplot as plt

file = "/p/tmp/yanleizh/newpypsa_0913/PyPSA-China-PIK/resources/derived_data/cutout_China-2020c_bare/availability_matrix_onwind.nc"

ds = xr.open_dataset(file)
da = ds["__xarray_dataarray_variable__"]

# 对所有 bus 求平均
da_mean = da.mean(dim="bus")

plt.figure(figsize=(8,6))
da_mean.plot(cmap="viridis")
plt.title("Mean Onshore Wind Availability (All Buses)", fontsize=16)
plt.show()
```
